```python
import pandas as pd
from sklearn.datasets import make_classification, make_moons
```

```python
X,y = make_moons(noise=0.3)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(
    x=X[:,0],
    y= X[:,1],
    c=y
)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
without_dropout_model = Sequential()
```

```python
without_dropout_model.add(Dense(128, input_dim = 2, activation= 'relu'))
without_dropout_model.add(Dense(128, activation= 'relu'))
without_dropout_model.add(Dense(1,activation='sigmoid'))
```

```python
without_dropout_model.compile(loss= 'binary_crossentropy', optimizer= 'adam', metrics= ['accuracy'])
```

```python
without_dropout_history = without_dropout_model.fit(X, y, epochs=500, validation_split = 0.2,verbose=1)
```

```python
plt.plot(without_dropout_history.history['loss'])
plt.plot(without_dropout_history.history['val_loss'])
plt.xlim(200)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X, y.astype('int'), clf=without_dropout_model, legend=2)
plt.show()
```

```python
dropout_model = Sequential()
```

```python
dropout_model.add(Dense(128,input_dim = 2, activation='relu'))
dropout_model.add(Dropout(0.5))
dropout_model.add(Dense(128,activation='relu'))
dropout_model.add(Dropout(0.5))
dropout_model.add(Dense(1,activation='sigmoid'))
```

```python
dropout_model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
dropout_history = dropout_model.fit(X, y, epochs=500, validation_split = 0.2,verbose=1)
```

```python
plt.plot(dropout_history.history['loss'])
plt.plot(dropout_history.history['val_loss'])
plt.xlim(200)
```

```python
plot_decision_regions(X, y.astype('int'), clf=dropout_model, legend=2)
plt.show()
```
