```python
import pandas as pd
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import L2, L1, L1L2
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(centers=2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2 , random_state= 56)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(
    x= X[:,0],
    y= X[:,1],
    c = y
)
```

```python
model = Sequential()
```

```python
model.add(Dense(128, input_dim =2 , activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=500)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X, y.astype('int'), clf=model, legend=2)
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

## **Regularization**

```python
regularized_model = Sequential()
```

```python
regularized_model.add(Dense(128, activation= 'relu', input_dim = 2, kernel_regularizer= L2(l2=0.03)))
regularized_model.add(Dense(128, activation= 'relu', kernel_regularizer= L2(l2=0.03)))
regularized_model.add(Dense(1, activation= 'sigmoid'))
```

```python
regularized_model.compile(loss='binary_crossentropy', optimizer='adam', metrics= ['accuracy'])
```

```python
regularized_history = regularized_model.fit(X_train, y_train , validation_data= (X_test, y_test), epochs= 500)
```

```python
plot_decision_regions(X, y.astype('int'), clf=regularized_model, legend=2)
plt.show()
```

```python
plt.plot(regularized_history.history['loss'])
plt.plot(regularized_history.history['val_loss'])
plt.xlim(0,200)
plt.ylim(0,2)
```
