## **Activation Functions**

### **Linear Activation Function Implementation**

```python
from mlxtend.plotting import plot_decision_regions
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt 
import keras.datasets.mnist as minst
```

```python
(X_train,y_train), (X_test, y_test) = minst.load_data()
```

```python
X_train.shape
```

```python
linear_model = Sequential([
    Flatten(input_shape= (28,28)),
    Dense(128, activation='linear'),
    Dense(10, activation='softmax')
])
```

```python
linear_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics= ['accuracy'])
```

#### **Scalling the inputs**

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
linear_history = linear_model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs= 10, batch_size= 256)
```

```python
plt.plot(linear_history.history['loss'])
plt.plot(linear_history.history['val_loss'])
```

```python
linear_model.evaluate(X_test,y_test)
```

### **Leaky Relu Activation functcion Implementation**

```python
from tensorflow.keras.layers import Input, Dense, LeakyReLU, Flatten, Softmax
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

```python
inputs = Input(shape=(28,28))
x = Flatten()(inputs)
x = Dense(128,name = 'dense_1')(x)
x = LeakyReLU(alpha=0.1)(x)
x = Dense(10, name='output')(x)
outputs = Softmax()(x)

model_leaky = Model(inputs,outputs)
```

```python
model_leaky.compile(
    loss = 'sparse_categorical_crossentropy',
    optimizer = Adam(learning_rate=0.01),
    metrics = ['accuracy']
)
```

```python
leaky_history = model_leaky.fit(
    X_train,
    y_train,
    epochs = 10,
    validation_split = 0.2,
    batch_size = 256
)
```

```python
model_leaky.evaluate(X_test,y_test)
```

### **Parametric Relu Activation Function Implementation**

```python
import tensorflow
from tensorflow.keras.layers import PReLU
```

```python
inputs = Input(shape=(28,28), name='input_layer')
x = Flatten(name = 'flatten_input_layer')(inputs)
x = Dense(128, name= 'dense_1')(x)
x = PReLU(alpha_initializer= tensorflow.keras.initializers.Constant(0.2),name = 'PRelu_activation_layer_1')(x)
x = Dense(10,name= 'pre_output_layer')(x)
outputs = Softmax(name = 'output_layer')(x)

model_prelu = Model(inputs,outputs)
```

```python
model_prelu.layers
```

```python
model_prelu.compile(
    loss= 'sparse_categorical_crossentropy', 
    optimizer = Adam(learning_rate=0.01),
    metrics = ['accuracy']
    )
```

```python
prelu_histroy = model_prelu.fit(
    X_train,
    y_train,
    epochs = 10,
    validation_split = 0.2,
    batch_size = 256
)
```

```python
model_prelu.evaluate(X_test,y_test)
```
