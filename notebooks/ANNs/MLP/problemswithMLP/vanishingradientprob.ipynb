```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=0.2, random_state=42)
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=X[:,0],y=X[:,1], c=y)
```

```python
model = Sequential()
```

```python
#Layers 

#Input Layer
model.add(Dense(10,activation='relu',input_dim = 2))

#Hidden Layers
for i in range(7):
    model.add(Dense(10,activation='relu'))
    
#Output layer
model.add(Dense(1,activation='relu'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics= ['accuracy'])
```

```python
old_weights = model.get_weights()[0]
```

```python
old_weights
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size=0.2, random_state=34)
```

```python
model.fit(X_train,y_train, epochs=100)
```

## **RELU**

```python
model_relu = Sequential()

model_relu.add(Dense(10,activation='relu', input_dim = 2))

for i in range(7):
    model_relu.add(Dense(10,activation='relu'))
    
model_relu.add(Dense(1,activation='relu'))
```

```python
model_relu.compile(loss='binary_crossentropy', optimizer='Adam', metrics= ['accuracy'])
```

```python
old_relu_weights = model_relu.get_weights()[0]
```

```python
old_relu_weights
```

```python
model_relu.fit(X_train,y_train,epochs=100)
```

```python
relu_new_weights = model_relu.get_weights()[0]
```

```python
gradient = (relu_new_weights - old_relu_weights) / 0.001
```
