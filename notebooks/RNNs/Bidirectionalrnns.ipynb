```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense,LSTM,GRU,Bidirectional,Dropout
```

```python
num_words = 10000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=num_words)
```

```python
maxlen = 100
X_train = sequence.pad_sequences(X_train, maxlen=maxlen,padding='post',truncating='post')
X_test = sequence.pad_sequences(X_test, maxlen=maxlen,padding='post',truncating='post')
```

```python
embedding_dim = 32
```

```python
bi_simple_rnn_model = Sequential()
bi_simple_rnn_model.add(Embedding(input_dim=num_words, output_dim=embedding_dim, input_length=maxlen,input_shape =(maxlen,)))
bi_simple_rnn_model.add(Bidirectional(SimpleRNN(units=5,dropout=0.2,recurrent_dropout=0.1)))
bi_simple_rnn_model.add(Dropout(0.2))
bi_simple_rnn_model.add(Dense(1, activation='sigmoid'))
```

```python
bi_simple_rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
bi_simple_rnn_model.summary()
```

```python
bi_simple_rnn_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

```python
bi_lstm_model = Sequential()
bi_lstm_model.add(Embedding(input_dim=num_words, output_dim=embedding_dim, input_length=maxlen,input_shape =(maxlen,)))
bi_lstm_model.add(Bidirectional(LSTM(units=5,dropout=0.2,recurrent_dropout=0.1)))
bi_lstm_model.add(Dropout(0.2))
bi_lstm_model.add(Dense(1, activation='sigmoid'))
```

```python
bi_lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
bi_lstm_model.summary()
```

```python
bi_lstm_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

```python
bi_gru_model = Sequential()
bi_gru_model.add(Embedding(input_dim=num_words, output_dim=embedding_dim,input_length=maxlen,input_shape =(maxlen,)))
bi_gru_model.add(Bidirectional(GRU(units=5,dropout=0.2,recurrent_dropout=0.1)))
bi_gru_model.add(Dropout(0.2))
bi_gru_model.add(Dense(1, activation='sigmoid'))
```

```python
bi_gru_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
bi_gru_model.summary()
```

```python
bi_gru_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```
