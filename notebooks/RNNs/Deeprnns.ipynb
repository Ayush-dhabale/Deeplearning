## **GRUs**

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense,LSTM,GRU
```

```python
(x_train,y_train),(x_test,y_test)=imdb.load_data(num_words=100000)
```

```python
x_train=sequence.pad_sequences(x_train,maxlen=500)
x_test=sequence.pad_sequences(x_test,maxlen=500)
```

```python
gru_model=Sequential()
gru_model.add(Embedding(input_dim=100000, output_dim=128, input_length=500,input_shape=(500,)))
gru_model.add(GRU(32,return_sequences=True))
gru_model.add(GRU(32,return_sequences=True))
gru_model.add(GRU(32))
gru_model.add(Dense(1,activation='sigmoid'))
```

```python
gru_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
gru_model.summary()
```

```python
gru_model.fit(x_train,y_train,epochs=10,batch_size=128,validation_data=(x_test,y_test))
```

## **LSTMs**

```python
lstm_model = Sequential()
lstm_model.add(Embedding(input_dim=100000, output_dim=128, input_length=500,input_shape=(500,)))
lstm_model.add(LSTM(32,return_sequences=True))
lstm_model.add(LSTM(32,return_sequences=True))
lstm_model.add(LSTM(32))
lstm_model.add(Dense(1,activation='sigmoid'))
```

```python
lstm_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
lstm_model.summary()
```

```python
lstm_model.fit(x_train,y_train,epochs=10,batch_size=128,validation_data=(x_test,y_test))
```
