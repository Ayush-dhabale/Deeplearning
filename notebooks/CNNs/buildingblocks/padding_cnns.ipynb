## **CNNs**

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Activation, Input
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

## **Valid Padding**

```python
input_layer = Input(shape=(28, 28, 1))
conv1 = Conv2D(32, kernel_size=(3, 3),padding='valid', activation='relu')(input_layer)
conv2 = Conv2D(64, kernel_size=(3, 3),padding='valid', activation='relu')(conv1)
conv3 = Conv2D(128, kernel_size=(3, 3),padding='valid', activation='relu')(conv2)
flat = Flatten()(conv3)
dense1 = Dense(128, activation='relu')(flat)
output_layer = Dense(10, activation='softmax')(dense1)

model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

## **Same Padding**

```python
input_layer = Input(shape=(28, 28, 1))
conv1 = Conv2D(32, kernel_size=(3, 3),padding='same', activation='relu')(input_layer)
conv2 = Conv2D(64, kernel_size=(3, 3),padding='same', activation='relu')(conv1)
conv3 = Conv2D(128, kernel_size=(3, 3),padding='same', activation='relu')(conv2)
flat = Flatten()(conv3)
dense1 = Dense(128, activation='relu')(flat)
output_layer = Dense(10, activation='softmax')(dense1)

model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```
